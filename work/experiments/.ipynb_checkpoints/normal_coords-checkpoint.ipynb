```python
def gen_selftrans(n):
    a = np.random.randn(n, n) * 5
    a = (a + a.T) / 2
    return a

def lower_triangle(n):
    a = np.random.randn(n, n) * 5
    for i in range(n):
        for j in range(i + 1, n):
            a[i, j] = 0
    return a

n = 5
a = gen_selftrans(n)
```
