# 1.

Функция, считающая TPR и FPR для всех возможных порогов

```python
def fpr_tpr(xs, ys):
    inds = np.argsort(xs)
    ys = ys[inds]
    rys = 1 - ys
    return np.cumsum(rys) / rys.sum(), np.cumsum(ys) / ys.sum()
```

Функции, извлекающая из всех возможных разбиений ROC кривую

```python
def make_transformation(vals, bools):
    return vals * (1 - bools) + (1 - vals) * bools

def optimize(fprs, tprs):
    to_rebuild = tprs < fprs
    fprs = make_transformation(fprs, to_rebuild)
    tprs = make_transformation(tprs, to_rebuild)
    inds = np.argsort(fprs)
    return fprs[inds], tprs[inds]
    
def roc(y_real, y_pred):
    return optimize(*fpr_tpr(y_pred, y_real))

def roc_auc(y_real, y_pred):
    xs, ys = roc(y_real, y_pred)
    return np.trapz(ys, xs)
```

```python
for i in range(3):
    plt.plot(*optimize(*fpr_tpr(X[:, -1 - i], y)))
plt.show()
```

Фунции, вычисляющие precision, recall и извлекающие из всего множество парето фронт

```python
def precision_recall_1(xs, ys):
    inds = np.argsort(xs)
    ys = ys[inds]
    rys = 1 - ys
    return np.cumsum(ys) / ys.sum(), np.cumsum(ys) / (np.arange(ys.shape[0]) + 1)

def precision_recall_2(xs, ys):
    inds = np.argsort(xs)
    ys = ys[inds[::-1]]
    rys = 1 - ys
    return np.cumsum(ys) / ys.sum(), np.cumsum(ys) / (np.arange(ys.shape[0]) + 1)

def precision_recall(xs, ys):
    rec1, prec1 = precision_recall_1(xs, ys)
    rec2, prec2 = precision_recall_2(xs, ys)
    return np.hstack([rec1, rec2]), np.hstack([prec1, prec2])

def pareto_frontier(xs, ys):
    res_x = []
    res_y = []
    for x, y in zip(xs, ys):
        if np.logical_or(np.logical_and(xs == x, ys == y), np.logical_or(xs < x, ys < y)).all():
            res_x.append(x)
            res_y.append(y)
    res_x = np.array(res_x)
    res_y = np.array(res_y)
    inds = np.argsort(res_x)
    return res_x[inds], res_y[inds]
```

```python
def plogp(sums, cnts):
    good = sums != 0
    
    first = good * (sums / cnts * np.log(sums / cnts + np.logical_not(good)))
    second = sums / cnts * np.log(sums / cnts + np.logical_not(good))
    assert (first == second).all()
    
    return first
        
def entropy(xs, ys, flag=False):
    inds = np.argsort(xs)
    
    if flag:
        print(xs)
        print(inds)
    
    x_sorted = xs[inds]
    y_sorted = ys[inds]
    
    n = ys.shape[0]
    y_sum = y_sorted.sum();
    
    unique, inds = np.unique(x_sorted, True)
    
    if inds.shape[0] == 1:
        return None, None
    inds = inds[1:] - 1
    
    a = np.cumsum(y_sorted)[inds]
    b = np.cumsum(1 - y_sorted)[inds]
    c = y_sum - a
    d = (n - y_sum) - b
    
    ab = a + b
    cd = c + d
    ac = a + c
    bd = b + d
    
#     hx = -ac / n * np.log(ac / n) - bd / n * np.log(bd / n)
#     h1 = -a / ab * np.log(a / ab) - b / ab * np.log(b / ab)
#     h2 = -c / cd * np.log(c / cd) - d / cd * np.log(d / cd)
    hx = -plogp(ac, n) - plogp(bd, n)
    h1 = -plogp(a, ab) - plogp(b, ab)
    h2 = -plogp(c, cd) - plogp(d, cd)

    gain = hx - ab / n * h1 - cd / n * h2

    best_ind = np.argmax(gain)
    if flag:
        print('best ind values: {} {} {}'.format(hx[best_ind], h1[best_ind], h2[best_ind]))

    return gain[best_ind], x_sorted[inds[best_ind]]
```

Классы дерева и функция построения

```python
class Leaf:
    def __init__(self, prob):
        self.prob = prob

    def __call__(self, X):
        return np.full(X.shape[0], self.prob)
    
class Vertex:
    def __init__(self, left, right, ind, threshold):
        self.left = left
        self.right = right
        self.ind = ind
        self.threshold = threshold
        
    def __call__(self, X):
        left_inds = X[:, self.ind] <= self.threshold
        right_inds = np.logical_not(left_inds)
        
        result = np.zeros(X.shape[0])
        result[left_inds] = self.left(X[left_inds])
        result[right_inds] = self.right(X[right_inds])
        
        return result
                
def random_seq(n):
    result = list(range(n))
    random.shuffle(result)
    return result
    
def build_tree(X, y, criteria, h):
    if h == 0 or y.sum() == y.shape[0] or y.sum() == 0:
        return Leaf(y.sum() / y.shape[0])
    else:
        best_gain = None
        threshold = None
        best_index = None
        
        for i in random_seq(X.shape[1]):
            gain, threshold = criteria(X[:, i], y)
            if gain is not None and (best_gain is None or gain > best_gain):
                best_gain = gain
                best_threshold = threshold
                best_index = i
                
        if best_index is None:
            return Leaf(y.sum() / y.shape[0])
                
        left_inds = X[:, best_index] <= best_threshold
        right_inds = np.logical_not(left_inds)
        
#         if not(left_inds.sum() and right_inds.sum()):    
#             print(left_inds.sum(), right_inds.sum())
#             print(best_gain, best_threshold, best_index)
#             criteria(X[:, best_index], y, True)
            
        assert left_inds.sum() and right_inds.sum()
        
        first_tree = build_tree(X[left_inds, :], y[left_inds], criteria, h - 1)
        second_tree = build_tree(X[right_inds, :], y[right_inds], criteria, h - 1)
        return Vertex(first_tree, second_tree, best_index, best_threshold)
```

Класс для храниения случайного леса и функция построения.

```python
class RandomForrest:
    def __init__(self, trees, indss, choose):
        self.trees = trees
        self.indss = indss
        self.choose = choose
    
    def predict(self, X):
        result = np.zeros((X.shape[0], len(self.trees)))
        for i, (tree, inds) in enumerate(zip(self.trees, self.indss)):
            result[:, i] = tree(X[:, inds])
        return self.choose(result)
    
def build_forrest(X, y, criteria, depth, tree_cnt, features_per_tree, choose):
    trees = []
    indss = []
    for i in range(tree_cnt):
        X_cur, y_cur = skl.utils.resample(X, y)
        inds = skl.utils.resample(np.arange(features_per_tree), replace=False)
        trees.append(build_tree(X_cur[:, inds], y_cur, criteria, depth))
        indss.append(inds)
    return RandomForrest(trees, indss, choose)
```

Три стратегии согласования результатов

```python
def choose_mean(result):
    return (result).sum(axis=1) / result.shape[1]
    
def choose_threshold(result):
    return (result > 0.5).sum(axis=1) / result.shape[1]

def choose_log(result):
    return np.log(result + 1e-5).sum(axis=1)
```

```python
sizes = np.arange(1, 101, 20)
aucs = [roc_auc(y_test, build_forrest(X, y, entropy, 5, size, X.shape[1], choose_log).predict(X_test)) for size in sizes]
plt.plot(sizes, aucs)
plt.xlabel('tree count')
plt.ylabel('roc auc score')
plt.show()
best_size = 100

depths = np.arange(1, 20, 3)
aucs = [roc_auc(y_test, build_forrest(X, y, entropy, depth, best_size, X.shape[1], choose_log).predict(X_test)) for depth in depths]
plt.plot(depths, aucs)
plt.xlabel('tree max depth')
plt.ylabel('roc auc score')
plt.show()
best_depth = 20

features = list(range(1, X.shape[1], 10)) + [X.shape[1]]
aucs = [roc_auc(y_test, build_forrest(X, y, entropy, best_depth, best_size, feature, choose_log).predict(X_test)) for feature in features]
plt.plot(features, aucs)
plt.xlabel('feature count')
plt.ylabel('roc auc score')
plt.show()
best_depth = X.shape[1]
```
