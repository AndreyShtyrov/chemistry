```python
# %matplotlib inline
%matplotlib notebook

from math import *

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
from matplotlib import cm
```

```python
with open('../func_backup_real.out') as f:
    exec(f.read())

def plot(xs, ys):
    plt.scatter(xs[0], ys[0], color='b')
    plt.scatter(xs[1], ys[1], color='r')
    plt.plot(xs, ys, 'b')

xs, ys = [], []
with open('../a.out', 'r') as f:
    for i in f:
        vals = i.strip().split()
        if vals:
            x, y = vals
            xs.append(float(x))
            ys.append(float(y))
        else:
            plot(xs, ys)
            xs, ys = [], []
    if xs:
        plot(xs, ys)
        xs, ys = [], []
```
