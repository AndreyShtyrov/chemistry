```python
def split_all(file):
    with open(file, 'r') as f:
        for line in f:
            yield from line.split()

def next_int(file_generator):
    return int(next(file_generator))
    
def next_float(file_generator):
    return float(next(file_generator))
    
    
def read_matrix(inp):
    n, m = next_int(inp), next_int(inp)
    vals = np.zeros((n, m))
    for i in range(n):
        for j in range(m):
            vals[i, j] = next_float(inp)
    return vals

N = 500
f = split_all('../func.out')
# a = read_matrix(f)
# print(a.shape)

# fig = plt.figure()
# ax = fig.gca(projection='3d')
# ax.plot_surface(a[:, 0].reshape(N, -1), a[:, 1].reshape(N, -1), a[:, 2].reshape(N, -1), cmap=cm.coolwarm)

aas = []
for i in range(10):
    aas.append(read_matrix(f))
paths = []
for i in range(10):
    paths.append(read_matrix(f))

for a, path in zip(aas, paths):
    fig = plt.figure()
    plt.plot(a[:, 0], a[:, 1], label='value')
    plt.plot(a[:, 0], a[:, 2], label='grad')
    plt.legend()



    plt.plot(path[:, 0], path[:, 1])
    plt.scatter(path[:, 0], path[:, 1], s=2)
    
    for i in range(1, len(path)):
        print(path[i, 0] - path[i, 1])
    
    break
    
    
```
